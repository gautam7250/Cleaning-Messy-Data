# Handling outliers

```python
# Droping the outliers using Standard Deviation
# Read the data
data = pd.read_csv('D:\Educational(B.E.)\Course\DATA1\Python-Data-Analysis-Third-Edition-master\Chapter07\employee.csv')
```

```python
data
```

```python
# Droping the outliers using Standard Deviation
upper_limit = data['performance_score'].mean() + 3 * data['performance_score'].std()
lower_limit = data['performance_score'].mean() - 3 * data['performance_score'].std()
```

```python
data = data[(data['performance_score']<upper_limit) & (data['performance_score']>lower_limit)]
```

# Feature Scaling

```python
# In real life, most features have different franges , magnitudes ,and units , such as age being between 0-20 and salary being betwwn 0 to thousands or misllions . From a data analyst or data scientis poient of vies, how can we compare these features when they are on different scales ? High-magnitude features will weigh more on machine learning models the lower magnitude features. Thankfully, feature scaling or feature normalize can solve such issues.

# Featue scaling brings all the featues to the same level of magnitude . This is not sompulsory for all kinds of alogotithms ; some alogorithms clearly need scaled data, such as those that rely on Eucliidean distance measures such as K-nearest neighbor and the k-means clustering algorithm.
```

# Feature transformation

```python
data = pd.read_csv('D:\Educational(B.E.)\Course\DATA1\Python-Data-Analysis-Third-Edition-master\Chapter07\employee.csv')
```

```python
# Create performnace grade functkion
def performance_grade(score):
    if score >= 700:
        return 'A'
    elif score < 700 and score >= 500:
        return 'B'
    else:
        return 'C'
# Apply performnace grade function on whole DataFrame using apply() function
data['performance_grade'] = data.performance_score.apply(performance_grade)
data.head()
```

```python
# Split the name column in first and last name
data['first_name']=data.name.str.split(" ").map(lambda var: var[0])
data['last_name']=data.name.str.split(" ").map(lambda var: var[1])
```
